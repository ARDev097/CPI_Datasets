```python
import pandas as pd
import os
from datetime import datetime
```

```python
# Define the folder containing the files
folder_path = "../Data/Delegates_Data_with_ACC_Members/"

# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
```

```python
# Load data from other CSV files (unchanged)
citizens_round2 = pd.read_csv("../Data/RPGF_Rounds_Data/Round 2.csv", encoding='latin1')
citizens_round3 = pd.read_csv("../Data/RPGF_Rounds_Data/Round 3.csv", encoding='latin1')
citizens_round4 = pd.read_csv("../Data/RPGF_Rounds_Data/Round 4.csv", encoding='latin1')
citizens_round5 = pd.read_csv("../Data/RPGF_Rounds_Data/Round 5.csv", encoding='latin1')
citizens_round6 = pd.read_csv("../Data/RPGF_Rounds_Data/Round 6.csv", encoding='latin1')

grants = pd.read_csv("../Data/Councils_and_Committees_Data/Grants_Council.csv", encoding='latin1')
grants_mm = pd.read_csv("../Data/Councils_and_Committees_Data/Grants_Council_MM.csv", encoding='latin1')
dab = pd.read_csv("../Data/Councils_and_Committees_Data/Developer_Advisory_Board.csv", encoding='latin1')
coc = pd.read_csv("../Data/Councils_and_Committees_Data/Code_of_Conduct_Council.csv", encoding='latin1')
```

```python
# Drop rows with null values in the 'address' column for each DataFrame
citizens_round2.dropna(subset=['address'], inplace=True)
citizens_round3.dropna(subset=['address'], inplace=True)
citizens_round4.dropna(subset=['address'], inplace=True)
citizens_round5.dropna(subset=['address'], inplace=True)
citizens_round6.dropna(subset=['address'], inplace=True)

grants.dropna(subset=['address'], inplace=True)
grants_mm.dropna(subset=['address'], inplace=True)
dab.dropna(subset=['address'], inplace=True)
coc.dropna(subset=['address'], inplace=True)
```

```python
# Filter grants data by season 
grants_season3 = grants[grants['season'] == 3]
grants_season4 = grants[grants['season'] == 4]
grants_season5 = grants[grants['season'] == 5]
grants_mm_season5 = grants_mm[grants_mm['season'] == 5]
dab_season5 = dab[dab["season"] == 5]
coc_season5 = coc[coc["season"] == 5]
grants_season6 = grants[grants['season'] == 6]
grants_mm_season6 = grants_mm[grants_mm['season'] == 6]
dab_season6 = dab[dab["season"] == 6]
coc_season6 = coc[coc["season"] == 6]
```

```python
# Convert start_date and end_date columns to datetime for comparison
citizens_round2.loc[:, 'start_date'] = pd.to_datetime(citizens_round2['start_date'], format='%d-%m-%Y')
citizens_round2.loc[:, 'end_date'] = pd.to_datetime(citizens_round2['end_date'], format='%d-%m-%Y')

citizens_round3.loc[:, 'start_date'] = pd.to_datetime(citizens_round3['start_date'], format='%d-%m-%Y')
citizens_round3.loc[:, 'end_date'] = pd.to_datetime(citizens_round3['end_date'], format='%d-%m-%Y')

citizens_round4.loc[:, 'start_date'] = pd.to_datetime(citizens_round4['start_date'], format='%d-%m-%Y')
citizens_round4.loc[:, 'end_date'] = pd.to_datetime(citizens_round4['end_date'], format='%d-%m-%Y')

citizens_round5.loc[:, 'start_date'] = pd.to_datetime(citizens_round5['start_date'], format='%d-%m-%Y')
citizens_round5.loc[:, 'end_date'] = pd.to_datetime(citizens_round5['end_date'], format='%d-%m-%Y')

citizens_round6.loc[:, 'start_date'] = pd.to_datetime(citizens_round6['start_date'], format='%d-%m-%Y')
citizens_round6.loc[:, 'end_date'] = pd.to_datetime(citizens_round6['end_date'], format='%d-%m-%Y')

grants_season3.loc[:, 'start_date'] = pd.to_datetime(grants_season3['start_date'], format='%d-%m-%Y')
grants_season3.loc[:, 'end_date'] = pd.to_datetime(grants_season3['end_date'], format='%d-%m-%Y')

grants_season4.loc[:, 'start_date'] = pd.to_datetime(grants_season4['start_date'], format='%d-%m-%Y')
grants_season4.loc[:, 'end_date'] = pd.to_datetime(grants_season4['end_date'], format='%d-%m-%Y')

grants_season5.loc[:, 'start_date'] = pd.to_datetime(grants_season5['start_date'], format='%d-%m-%Y')
grants_season5.loc[:, 'end_date'] = pd.to_datetime(grants_season5['end_date'], format='%d-%m-%Y')

grants_mm_season5.loc[:, 'start_date'] = pd.to_datetime(grants_mm_season5['start_date'], format='%d-%m-%Y')
grants_mm_season5.loc[:, 'end_date'] = pd.to_datetime(grants_mm_season5['end_date'], format='%d-%m-%Y')

dab_season5.loc[:, 'start_date'] = pd.to_datetime(dab_season5['start_date'], format='%d-%m-%Y')
dab_season5.loc[:, 'end_date'] = pd.to_datetime(dab_season5['end_date'], format='%d-%m-%Y')

coc_season5.loc[:, 'start_date'] = pd.to_datetime(coc_season5['start_date'], format='%d-%m-%Y')
coc_season5.loc[:, 'end_date'] = pd.to_datetime(coc_season5['end_date'], format='%d-%m-%Y')

grants_season6.loc[:, 'start_date'] = pd.to_datetime(grants_season6['start_date'], format='%d-%m-%Y')
grants_season6.loc[:, 'end_date'] = pd.to_datetime(grants_season6['end_date'], format='%d-%m-%Y')

grants_mm_season6.loc[:, 'start_date'] = pd.to_datetime(grants_mm_season6['start_date'], format='%d-%m-%Y')
grants_mm_season6.loc[:, 'end_date'] = pd.to_datetime(grants_mm_season6['end_date'], format='%d-%m-%Y')

dab_season6.loc[:, 'start_date'] = pd.to_datetime(dab_season6['start_date'], format='%d-%m-%Y')
dab_season6.loc[:, 'end_date'] = pd.to_datetime(dab_season6['end_date'], format='%d-%m-%Y')

coc_season6.loc[:, 'start_date'] = pd.to_datetime(coc_season6['start_date'], format='%d-%m-%Y')
coc_season6.loc[:, 'end_date'] = pd.to_datetime(coc_season6['end_date'], format='%d-%m-%Y')
```

```python
# Helper function to add missing delegates
def add_missing_delegates(data, new_addresses):
    missing_addresses = set(new_addresses['address'].str.lower()) - set(data['delegate'].str.lower())
    missing_df = pd.DataFrame({
        'delegate': list(missing_addresses),
        'voting_power': 0
    })
    data = pd.concat([data, missing_df], ignore_index=True)
    return data
```

```python
# Process each file in the folder
for file_name in csv_files:
    file_path = os.path.join(folder_path, file_name)
    
    # Load the daily delegate data
    data = pd.read_csv(file_path)

    # Extract the date from the file name (assumed format 'YYYY-MM-DD.csv')
    file_date = datetime.strptime(file_name[:-4], '%Y-%m-%d')

    # Add missing delegates from various rounds and councils (unchanged)
    data = add_missing_delegates(data, citizens_round2)
    data = add_missing_delegates(data, citizens_round3)
    data = add_missing_delegates(data, citizens_round4)
    data = add_missing_delegates(data, citizens_round5)
    data = add_missing_delegates(data, citizens_round6)
    data = add_missing_delegates(data, grants_season3)
    data = add_missing_delegates(data, grants_season4)
    data = add_missing_delegates(data, grants_season5)
    data = add_missing_delegates(data, grants_mm_season5)
    data = add_missing_delegates(data, dab_season5)
    data = add_missing_delegates(data, coc_season5)
    data = add_missing_delegates(data, grants_season6)
    data = add_missing_delegates(data, grants_mm_season6)
    data = add_missing_delegates(data, dab_season6)
    data = add_missing_delegates(data, coc_season6)
    
    # Add the columns (unchanged)
    data['th_vp'] = None
    data['ch_member_r2'] = None
    data['ch_vp_r2'] = None
    data['ch_member_r3'] = None
    data['ch_vp_r3'] = None
    data['ch_member_r4'] = None
    data['ch_vp_r4'] = None
    data['ch_member_r5'] = None
    data['ch_vp_r5'] = None
    data['ch_member_r6'] = None
    data['ch_vp_r6'] = None
    data['gc_member_s3'] = None
    data['gc_vp_s3'] = None
    data['gc_member_s4'] = None
    data['gc_vp_s4'] = None
    data['gc_member_s5'] = None
    data['gc_vp_s5'] = None
    data['gc_member_mm_s5'] = None
    data['gc_vp_mm_s5'] = None
    data['sc_member_s5'] = None
    data['sc_vp_s5'] = None
    data['coc_member_s5'] = None
    data['coc_vp_s5'] = None
    data['dab_member_s5'] = None
    data['dab_vp_s5'] = None
    data['gc_member_s6'] = None
    data['gc_vp_s6'] = None
    data['gc_member_mm_s6'] = None
    data['gc_vp_mm_s6'] = None
    data['sc_member_s6'] = None
    data['sc_vp_s6'] = None
    data['coc_member_s6'] = None
    data['coc_vp_s6'] = None
    data['dab_member_s6'] = None
    data['dab_vp_s6'] = None

    # Create membership columns for existing delegates only if the round/council is active on the file_date
    data['ch_member_r2'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round2['start_date'].iloc[0] and 
                  file_date <= citizens_round2['end_date'].iloc[0] and
                  x.lower() in citizens_round2['address'].str.lower().values else 0)

    data['ch_member_r3'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round3['start_date'].iloc[0] and 
                  file_date <= citizens_round3['end_date'].iloc[0] and
                  x.lower() in citizens_round3['address'].str.lower().values else 0)

    data['ch_member_r4'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round4['start_date'].iloc[0] and 
                  file_date <= citizens_round4['end_date'].iloc[0] and
                  x.lower() in citizens_round4['address'].str.lower().values else 0)
    
    data['ch_member_r5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round5['start_date'].iloc[0] and 
                  file_date <= citizens_round5['end_date'].iloc[0] and
                  x.lower() in citizens_round5['address'].str.lower().values else 0)
    
    data['ch_member_r6'] = data['delegate'].apply(
        lambda x: 1 if file_date >= citizens_round6['start_date'].iloc[0] and 
                  file_date <= citizens_round6['end_date'].iloc[0] and
                  x.lower() in citizens_round6['address'].str.lower().values else 0)

    data['gc_member_s3'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_season3['start_date'].iloc[0] and 
                  file_date <= grants_season3['end_date'].iloc[0] and
                  x.lower() in grants_season3['address'].str.lower().values else 0)

    data['gc_member_s4'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_season4['start_date'].iloc[0] and 
                  file_date <= grants_season4['end_date'].iloc[0] and
                  x.lower() in grants_season4['address'].str.lower().values else 0)

    data['gc_member_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_season5['start_date'].iloc[0] and 
                  file_date <= grants_season5['end_date'].iloc[0] and
                  x.lower() in grants_season5['address'].str.lower().values else 0)

    data['gc_member_mm_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_mm_season5['start_date'].iloc[0] and 
                  file_date <= grants_mm_season5['end_date'].iloc[0] and
                  x.lower() in grants_mm_season5['address'].str.lower().values else 0)

    data['dab_member_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= dab_season5['start_date'].iloc[0] and 
                  file_date <= dab_season5['end_date'].iloc[0] and
                  x.lower() in dab_season5['address'].str.lower().values else 0)

    data['coc_member_s5'] = data['delegate'].apply(
        lambda x: 1 if file_date >= coc_season5['start_date'].iloc[0] and 
                  file_date <= coc_season5['end_date'].iloc[0] and
                  x.lower() in coc_season5['address'].str.lower().values else 0)
    
    data['gc_member_s6'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_season6['start_date'].iloc[0] and 
                  file_date <= grants_season6['end_date'].iloc[0] and
                  x.lower() in grants_season6['address'].str.lower().values else 0)

    data['gc_member_mm_s6'] = data['delegate'].apply(
        lambda x: 1 if file_date >= grants_mm_season6['start_date'].iloc[0] and 
                  file_date <= grants_mm_season6['end_date'].iloc[0] and
                  x.lower() in grants_mm_season6['address'].str.lower().values else 0)

    data['dab_member_s6'] = data['delegate'].apply(
        lambda x: 1 if file_date >= dab_season6['start_date'].iloc[0] and 
                  file_date <= dab_season6['end_date'].iloc[0] and
                  x.lower() in dab_season6['address'].str.lower().values else 0)

    data['coc_member_s6'] = data['delegate'].apply(
        lambda x: 1 if file_date >= coc_season6['start_date'].iloc[0] and 
                  file_date <= coc_season6['end_date'].iloc[0] and
                  x.lower() in coc_season6['address'].str.lower().values else 0)

    # Calculate and assign voting power percentages based on active membership
    sum_th_vp = data[data['voting_power'] > 1]['voting_power'].sum()
    data['th_vp'] = data.apply(lambda row: (row['voting_power'] * 100) / sum_th_vp if row['voting_power'] > 1 else 0, axis=1)

    count_ch_member_r2 = data['ch_member_r2'].sum()
    data['ch_vp_r2'] = data.apply(lambda row: (row['ch_member_r2'] * 100) / count_ch_member_r2 if row['ch_member_r2'] == 1 else 0, axis=1)

    count_ch_member_r3 = data['ch_member_r3'].sum()
    data['ch_vp_r3'] = data.apply(lambda row: (row['ch_member_r3'] * 100) / count_ch_member_r3 if row['ch_member_r3'] == 1 else 0, axis=1)

    count_ch_member_r4 = data['ch_member_r4'].sum()
    data['ch_vp_r4'] = data.apply(lambda row: (row['ch_member_r4'] * 100) / count_ch_member_r4 if row['ch_member_r4'] == 1 else 0, axis=1)

    count_ch_member_r5 = data['ch_member_r5'].sum()
    data['ch_vp_r5'] = data.apply(lambda row: (row['ch_member_r5'] * 100) / count_ch_member_r5 if row['ch_member_r5'] == 1 else 0, axis=1)

    count_ch_member_r6 = data['ch_member_r6'].sum()
    data['ch_vp_r6'] = data.apply(lambda row: (row['ch_member_r6'] * 100) / count_ch_member_r6 if row['ch_member_r6'] == 1 else 0, axis=1)

    count_gc_member_s3 = data['gc_member_s3'].sum()
    data['gc_vp_s3'] = data.apply(lambda row: (row['gc_member_s3'] * 100) / count_gc_member_s3 if row['gc_member_s3'] == 1 else 0, axis=1)

    count_gc_member_s4 = data['gc_member_s4'].sum()
    data['gc_vp_s4'] = data.apply(lambda row: (row['gc_member_s4'] * 100) / count_gc_member_s4 if row['gc_member_s4'] == 1 else 0, axis=1)

    count_gc_member_s5 = data['gc_member_s5'].sum()
    data['gc_vp_s5'] = data.apply(lambda row: (row['gc_member_s5'] * 100) / count_gc_member_s5 if row['gc_member_s5'] == 1 else 0, axis=1)

    count_gc_member_mm_s5 = data['gc_member_mm_s5'].sum()
    data['gc_vp_mm_s5'] = data.apply(lambda row: (row['gc_member_mm_s5'] * 100) / count_gc_member_mm_s5 if row['gc_member_mm_s5'] == 1 else 0, axis=1)

    count_dab_member_s5 = data['dab_member_s5'].sum()
    data['dab_vp_s5'] = data.apply(lambda row: (row['dab_member_s5'] * 100) / count_dab_member_s5 if row['dab_member_s5'] == 1 else 0, axis=1)

    count_coc_member_s5 = data['coc_member_s5'].sum()
    data['coc_vp_s5'] = data.apply(lambda row: (row['coc_member_s5'] * 100) / count_coc_member_s5 if row['coc_member_s5'] == 1 else 0, axis=1)

    count_gc_member_s6 = data['gc_member_s6'].sum()
    data['gc_vp_s6'] = data.apply(lambda row: (row['gc_member_s6'] * 100) / count_gc_member_s6 if row['gc_member_s6'] == 1 else 0, axis=1)

    count_gc_member_mm_s6 = data['gc_member_mm_s6'].sum()
    data['gc_vp_mm_s6'] = data.apply(lambda row: (row['gc_member_mm_s6'] * 100) / count_gc_member_mm_s6 if row['gc_member_mm_s6'] == 1 else 0, axis=1)

    count_dab_member_s6 = data['dab_member_s6'].sum()
    data['dab_vp_s6'] = data.apply(lambda row: (row['dab_member_s6'] * 100) / count_dab_member_s6 if row['dab_member_s6'] == 1 else 0, axis=1)

    count_coc_member_s6 = data['coc_member_s6'].sum()
    data['coc_vp_s6'] = data.apply(lambda row: (row['coc_member_s6'] * 100) / count_coc_member_s6 if row['coc_member_s6'] == 1 else 0, axis=1)
    
    output_folder_path = "../Data/Data_Sheets/"  # Location of daily data sheets
    output_file_path = os.path.join(output_folder_path, file_name)

    # Check if the output folder exists, if not, create it
    if not os.path.exists(output_folder_path):
        os.makedirs(output_folder_path)

    pd.set_option('future.no_silent_downcasting', True)
    data.fillna(0, inplace=True)

    # Save the updated data back to the same file (overwrite)
    data.to_csv(output_file_path, index=False)
                                                                                                                                                
    # Print confirmation message for each file
    print(f"File {file_name} updated successfully!")

print("All files updated successfully!")
```
